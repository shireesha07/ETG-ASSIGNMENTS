```python
prod_id=str(input("enter product id:"))
name=str(input("enter name:"))
price=int(input("enter price:"))
quantity=int(input("enter quantity:"))
discount=int(input("eneter discount:"))
expiry_date=int(input("eneter expiry_date"))

items[prod_id ]={'name':name,'price':price,'quantity':quantity,'discount':discount,'expiry_date':expiry_date}
js=json.dumps(items)
fd=open("items.json",'w')
fd.write(js)
fd.close()
```
